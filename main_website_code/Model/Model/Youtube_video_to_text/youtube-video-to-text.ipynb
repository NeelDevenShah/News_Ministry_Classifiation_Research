```python
import subprocess

def convert_mp4_to_wav(input_file, output_file):

    command = ["ffmpeg", "-i", input_file, "-vn", "-acodec", "pcm_s16le", output_file]
    subprocess.run(command, check=True)
    subprocess.run(["rm", "/kaggle/working/tmp/output_file.mp4"], capture_output=True)
    
convert_mp4_to_wav("/kaggle/working/tmp/output_file.mp4", "/kaggle/working/tmp/output_file.wav")
print("MP4 file converted to WAV file successfully!")
```

```python
import wave

def get_wav_file_time(wav_file_path):

    with wave.open(wav_file_path, "rb") as wav_file:
        sample_rate = wav_file.getframerate()
        num_samples = wav_file.getnframes()
        duration = num_samples / sample_rate
        return duration
    
duration = get_wav_file_time("/kaggle/working/tmp/output_file.wav")
print(f"The duration of the WAV file is: {duration:.2f} seconds.")
```

```python
def divide_wav_file_into_n_parts(input_file_path, output_file_prefix, n):
    with wave.open(input_file_path, "rb") as input_file:
        n_frames = input_file.getnframes()
        frame_rate = input_file.getframerate()
        frames_per_part = n_frames // n

        for i in range(n):
            start_frame = i * frames_per_part
            end_frame = (i + 1) * frames_per_part
            output_file_path = f"{output_file_prefix}_{i}.wav"

            with wave.open(output_file_path, "wb") as output_file:
                output_file.setparams(input_file.getparams())
                output_file.setnframes(end_frame - start_frame)

                frames = input_file.readframes(end_frame - start_frame)
                output_file.writeframes(frames)
        subprocess.run(["rm", "/kaggle/working/tmp/output_file.wav"], capture_output=True)
        
divide_wav_file_into_n_parts("/kaggle/working/tmp/output_file.wav", "/kaggle/working/tmp/parted", int(duration/67.84))
print("WAV file divided into N parts successfully!")
```

```python
from pytube import YouTube
import subprocess
import wave
import speech_recognition as sr

def download_audio(video_link):
    try:
        video = YouTube(video_link)
        audio = video.streams.filter(
            only_audio=True, file_extension='mp4').first()
        audio.download("/kaggle/working/tmp/", "output_file.mp4")
        print('Download Completed!')
    except:
        print("Connection Error")


def convert_mp4_to_wav():

    command = ["ffmpeg", "-i", "/kaggle/working/tmp/output_file.mp4", "-vn",
               "-acodec", "pcm_s16le", "/kaggle/working/tmp/output_file.wav"]
    subprocess.run(command, check=True)
    subprocess.run(["rm", "/kaggle/working/tmp/output_file.mp4"],
                   capture_output=True)


def get_wav_file_time():

    with wave.open("/kaggle/working/tmp/output_file.wav", "rb") as wav_file:
        sample_rate = wav_file.getframerate()
        num_samples = wav_file.getnframes()
        duration = num_samples / sample_rate
        return duration


def divide_wav_file_into_n_parts(n):

    input_file_path = "/kaggle/working/tmp/output_file.wav"
    output_file_prefix = "/kaggle/working/tmp/parted"

    with wave.open(input_file_path, "rb") as input_file:
        n_frames = input_file.getnframes()
        frame_rate = input_file.getframerate()
        frames_per_part = n_frames // n

        for i in range(n):
            start_frame = i * frames_per_part
            end_frame = (i + 1) * frames_per_part
            output_file_path = f"{output_file_prefix}_{i}.wav"
            
            with wave.open(output_file_path, "wb") as output_file:
                output_file.setparams(input_file.getparams())
                output_file.setnframes(end_frame - start_frame)

                frames = input_file.readframes(end_frame - start_frame)
                output_file.writeframes(frames)
        subprocess.run(["rm", "/kaggle/working/tmp/output_file.wav"], capture_output=True)


def printFiles(duration):
    final_text = ""
    r = sr.Recognizer()
    for i in range(int(duration/67.84)):
        with sr.AudioFile(f'/kaggle/working/tmp/parted_{i}.wav') as source:
            audio = r.record(source)
            try:
                text = r.recognize_google(audio)
                final_text += text
                subprocess.run(
                    ["rm", f"/kaggle/working/tmp/parted_{i}.wav"], capture_output=True)
            except TypeError:
                print(TypeError)
    return final_text


def predict(link):

    download_audio(link)
    convert_mp4_to_wav()
    duration = get_wav_file_time()
    divide_wav_file_into_n_parts(int(duration/67.84))
    response_text = printFiles(duration)
    return response_text
```
