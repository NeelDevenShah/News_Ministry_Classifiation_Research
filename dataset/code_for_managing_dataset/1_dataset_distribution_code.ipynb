Now merging all the three datatframes

```python
df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
for i in range(0, len(df)):
    df.iloc[i, 1] = df.iloc[i, 1][:550]
```

```python
df.head()
```

```python
df.describe(include='object')
```

```python
df = df.drop_duplicates(subset='title')
```

```python
df.insert(0, 'labels', '')
```

```python
df.describe(include='object')
```

```python
df.head()
```
